```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
t2 = np.array([48, 19.9, 27.4, 34.2])
v2 = np.array([190.44, 175.63, 179.87, 184.10])
```

```python
plt.scatter(t2, v2)
```

```python
slope, intercept, r_value, p_value, std_err = linregress(t2, v2)
print(f"Fitted Slope: {slope}")
print(f"Fitted Intercept: {intercept}")
print(f"x-intercept: {-intercept/slope}")
plt.scatter(t2, v2, label='Data', color='blue')
plt.plot(t2, slope * t2 + intercept, label='Linear Fit', color='red')
plt.xlabel('t2')
plt.ylabel('v2')
plt.title("v2 vs t2")
plt.legend()
plt.show()
```
