```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
wavelength = np.array([462, 533, 618]) #blue, green, red (nm)
wavelength_error = np.array([10, 10, 10])

stopping_voltage_mV = np.array([-903, -644, -290]) #blue, green, red (mV)
stopping_voltage_error_mV = np.array([15, 15, 15])

e = 1.602e-19
c = 3e8

max_KE = e * stopping_voltage_mV
max_KE_error = e * stopping_voltage_error

frequency = c / wavelength
frequency_error = c * wavelength_error / wavelength**2
```

```python
def linear_fit(frequency, h, phi):
    return h * frequency - phi

popt, pcov = curve_fit(linear_fit, frequency, max_KE, sigma=stopping_voltage_error)
h_fitted, phi_fitted = popt
h_error, phi_error = np.sqrt(np.diag(pcov))
```

```python
# Convert stopping voltage to volts (V)
stopping_voltage = stopping_voltage_mV / 1000  # in Volts
stopping_voltage_error = stopping_voltage_error_mV / 1000  # in Volts

# Constants
e = 1.602e-19  # elementary charge in Coulombs
c = 3e8  # speed of light in m/s

# Convert stopping voltage to maximum kinetic energy (in J)
max_KE = e * np.abs(stopping_voltage)  # in Joules
max_KE_error = e * stopping_voltage_error  # in Joules

# Convert wavelength to frequency
frequency = c / (wavelength * 1e-9)  # converting nm to meters
frequency_error = c / (wavelength_error * 1e-9)  # converting nm to meters

# Linear fit function for frequency vs max KE
def linear_fit(frequency, h, phi):
    return h * frequency - phi

# Perform curve fitting
popt, pcov = curve_fit(linear_fit, frequency, max_KE, sigma=max_KE_error)
h_fitted, phi_fitted = popt
h_error, phi_error = np.sqrt(np.diag(pcov))

# Convert Planck's constant to eVs
h_fitted_eV = h_fitted / e  # Convert from Js to eVs
# Convert work function to eV
phi_fitted_eV = phi_fitted / e  # Convert from J to eV

# Calculate cutoff frequency (in Hz)
cutoff_frequency = phi_fitted / h_fitted

# Calculate errors for cutoff frequency
cutoff_frequency_error = cutoff_frequency * np.sqrt((phi_error / phi_fitted)**2 + (h_error / h_fitted)**2)

# Generate a linspace for frequency for the fit
frequency_fit = np.linspace(np.min(frequency), np.max(frequency), 100)

# Calculate the corresponding maximum kinetic energy for the fit
max_KE_fit = linear_fit(frequency_fit, *popt)

# Enhanced plot data and fit with error bars
plt.errorbar(frequency, max_KE / e, yerr=max_KE_error / e, fmt='o', color='royalblue', 
             label='Data with error', capsize=5, markersize=6, elinewidth=1.5)  # Data with thicker error bars
plt.plot(frequency_fit, max_KE_fit / e, color='darkred', linestyle='-', linewidth=2, label=f'y = {h_fitted_eV:.4e} + {phi_fitted_eV:.4e}')  # Fit line

# Labels and title with larger font sizes
plt.xlabel('Frequency (Hz)', fontsize=12)
plt.ylabel('Max KE (eV)', fontsize=12)
plt.title('Maximum Kinetic Energy vs. Frequency', fontsize=14)

# Grid and legend
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)
plt.show()
# Results
print(f'Estimated Plancks constant (h) in eVs: {h_fitted_eV:.4e}  {h_error / e:.4e} eVs')
print(f'Estimated work function (phi) in eV: {phi_fitted_eV:.4e}  {phi_error / e:.4e} eV')
print(f'Estimated cutoff frequency (f): {cutoff_frequency:.4e}  {cutoff_frequency_error:.4e} Hz')
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Given values
wavelength = np.array([405, 532])  # wavelengths in nm
wavelength_error = np.array([10, 10])  # wavelength error in nm

stopping_voltage_mV = np.array([-1250, -300])  # stopping voltages in mV
stopping_voltage_error_mV = np.array([10, 10])  # stopping voltage error in mV

# Constants
e = 1.602e-19  # elementary charge in Coulombs
c = 3e8  # speed of light in m/s

# Convert stopping voltage to Volts (V)
stopping_voltage = stopping_voltage_mV / 1000  # in Volts
stopping_voltage_error = stopping_voltage_error_mV / 1000  # in Volts

# Calculate maximum kinetic energy (in J)
max_KE = e * np.abs(stopping_voltage)  # in Joules
max_KE_error = e * stopping_voltage_error  # in Joules

# Convert wavelength to frequency
wavelength_m = wavelength * 1e-9  # convert nm to meters
wavelength_error_m = wavelength_error * 1e-9  # convert nm to meters
frequency = c / wavelength_m  # frequency in Hz
frequency_error = c * wavelength_error_m / wavelength_m**2  # frequency error in Hz

# Linear fit function for frequency vs max KE
def linear_fit(frequency, h, phi):
    return h * frequency - phi

# Perform curve fitting with sigma as the absolute max_KE_error
popt, pcov = curve_fit(linear_fit, frequency, max_KE, sigma=max_KE_error, absolute_sigma=True)
h_fitted, phi_fitted = popt
h_error, phi_error = np.sqrt(np.diag(pcov))

# Convert Planck's constant to eVs
h_fitted_eV = h_fitted / e  # Convert from Js to eVs
# Convert work function to eV
phi_fitted_eV = phi_fitted / e  # Convert from J to eV

# Calculate cutoff frequency (in Hz)
cutoff_frequency = phi_fitted / h_fitted
cutoff_frequency_error = cutoff_frequency * np.sqrt((phi_error / phi_fitted)**2 + (h_error / h_fitted)**2)

# Generate a linspace for frequency for the fit
frequency_fit = np.linspace(np.min(frequency), np.max(frequency), 100)
max_KE_fit = linear_fit(frequency_fit, *popt)

# Plot with error bars
plt.errorbar(frequency, max_KE / e, yerr=max_KE_error / e, fmt='o', color='royalblue', 
             label='Data with error', capsize=5, markersize=6, elinewidth=1.5)
plt.plot(frequency_fit, max_KE_fit / e, color='red', linestyle='-', linewidth=2,
         label=f'y = {h_fitted_eV:.4e} + {phi_fitted_eV:.4e}')
plt.xlabel('Frequency (Hz)', fontsize=12)
plt.ylabel('Max KE (eV)', fontsize=12)
plt.title('Maximum Kinetic Energy vs. Frequency')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)
plt.show()

# Results
print(f'Estimated Plancks constant (h) in eVs: {h_fitted_eV:.4e}  {h_error / e:.4e} eVs')
print(f'Estimated work function (phi) in eV: {phi_fitted_eV:.4e}  {phi_error / e:.4e} eV')
print(f'Estimated cutoff frequency (f): {cutoff_frequency:.4e}  {cutoff_frequency_error:.4e} Hz')
```

```python
# Calculate the expected (fitted) max KE values at observed frequencies
max_KE_expected = linear_fit(frequency, *popt)

# Calculate chi-squared
chi2 = np.sum(((max_KE - max_KE_expected) / max_KE_error) ** 2)

# Degrees of freedom: number of data points - number of parameters in fit
degrees_of_freedom = len(frequency) - len(popt)

# Reduced chi-squared.
reduced_chi2 = chi2 / degrees_of_freedom

print(f"Chi-squared (): {chi2:.4f}")
print(f"Degrees of freedom: {degrees_of_freedom}")
print(f"Reduced Chi-squared (_): {reduced_chi2:.4f}")
```
