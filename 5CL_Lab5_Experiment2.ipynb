```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

## Data

```python
wavelength_nm = np.array([435.41, 484.16, 654.48])
wavelength_error_nm = np.array([9.51, 9.40, 8.91])
wavelength = wavelength_nm / 10**9
wavelength_error = wavelength_error_nm / 10**9

wavelength_reciprocal = 1 / wavelength
wavelength_reciprocal_error = wavelength_error / wavelength**2

n_i_reciprocal_squared = np.array([1/25,1/16,1/9]) 
```

## Scatter Plot

Scatter plot of $\frac{1}{\lambda}$ vs. $\frac{1}{n_i^2}$

```python
plt.errorbar(wavelength_reciprocal, n_i_reciprocal_squared, xerr=wavelength_reciprocal_error, fmt='o', ecolor='red', capsize=5)
plt.xlabel(r'$\frac{1}{\lambda}$ (nm⁻¹)')
plt.ylabel(r'$\frac{1}{n_i^2}$')
plt.title(r'$\frac{1}{\lambda}$ vs $\frac{1}{n_i^2}$')
plt.show()
```

## Fit

```python
def linearized_function(n_i, R):
    wavelength_reciprocal = R*(n_i - 1/4)
    return wavelength_reciprocal

popt, pcov = curve_fit(linearized_function, n_i_reciprocal_squared, wavelength_reciprocal)

R_fitted = popt[0]
R_uncertainty = np.sqrt(np.diag(pcov))[0]

n_i_fit = np.linspace(min(n_i_reciprocal_squared), max(n_i_reciprocal_squared), 100)
wavelength_reciprocal_fit = linearized_function(n_i_fit, R_fitted)

plt.errorbar(n_i_reciprocal_squared, wavelength_reciprocal, yerr=wavelength_reciprocal_error, fmt='o', ecolor='red', capsize=3, label='Data')
plt.plot(n_i_fit, wavelength_reciprocal_fit, label=f'Fit: R = {-1 * R_fitted:.4f}', color='blue')
plt.xlabel(r'$\frac{1}{n_i^2}$')
plt.ylabel(r'$\frac{1}{\lambda}$ (m⁻¹)')
plt.title(r'$\frac{1}{\lambda}$ vs $\frac{1}{n_i^2}$')
plt.legend()
plt.show()

print(f'R: {-1 * R_fitted:.0f} +/- {R_uncertainty:.0f}')
```
