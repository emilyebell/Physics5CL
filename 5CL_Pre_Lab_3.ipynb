```python
import numpy as np
from scipy.optimize import fsolve

def sin_error(theta):
    return abs(theta - np.sin(theta)) / np.sin(theta) - 0.01

def tan_error(theta):
    return abs(theta - np.tan(theta)) / np.tan(theta) - 0.01

def sin_tan_error(theta):
    return abs(np.sin(theta) - np.tan(theta)) / np.tan(theta) - 0.01

theta_sin = fsolve(sin_error, 0.1)[0]
theta_tan = fsolve(tan_error, 0.1)[0]
theta_sin_tan = fsolve(sin_tan_error, 0.1)[0]

theta_sin, theta_tan, theta_sin_tan #rad
```

```python
import numpy as np
import matplotlib.pyplot as plt

L = 1.00  # m
a = 1      # ref unit (=1)
d = 4 * a  
wavelength = a/125
y = np.linspace(-0.02, 0.02, 1000)  # range of positions (in meters), y| -2 cm <= y <= 2 cm
y_cm = y*100

# functionsss (assuming small angle approx: sin(theta) = y/L)
def single_slit_intensity(a, L, wavelength, y):
    alpha = (np.pi * a * y) / (L * wavelength)
    return (np.sin(alpha) / alpha)**2

def double_slit_intensity(a, d, L, wavelength, y):
    beta = (np.pi * d * y) / (L * wavelength)
    return (np.cos(beta))**2

I_single = single_slit_intensity(a, L, wavelength, y)
I_double = double_slit_intensity(a, d, L, wavelength, y)
I_combined = I_single * I_double  # total intensity

# Plotting

plt.plot(y_cm, I_combined, label="Double-slit", color='r') 

plt.plot(y_cm, I_single, label="Single-slit envelope", color='b')

plt.title("Intensity vs Position")
plt.xlabel("Position y (cm)")
plt.ylabel("Intensity I (enter units)")
plt.axvline(x=0, color='gray', linestyle='--')
plt.legend()
plt.axvline(x=0.64, color='g', linestyle='--', label="Interference Maxima")
plt.axvline(x=-0.64, color='g', linestyle='--')
plt.axvline(x=1.28, color='g', linestyle='--')
plt.axvline(x=-1.28, color='g', linestyle='--')
plt.axvline(x=0.96, color='purple', linestyle='--', label="Diffraction Minima")
plt.axvline(x=-0.96, color='purple', linestyle='--')
plt.legend()

plt.show()
```
